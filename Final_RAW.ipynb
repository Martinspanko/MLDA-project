## 4. Model selection <a name='model_selection'></a>

### Libraries
Importing libraries for model selection

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV

# Algorithmns models to be compared
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

### Classification metric
Before we can compare classifiers we need to decide on a classification metric. We have learned about several metrics in class. A common metric used is the accuracy score, which is the opposite of the error rate. The accuracy score calculates how often the prediction is correct. Two other common metrics are precision and recall. These metrics provide information about the false positive and false negative predictions. And finally, the perhaps most commonly used metric is the f1 score, which is the harmonic mean of precision and recall and gives a better measure of the incorrectly classified cases than the accuracy metric. The f1-score is especially usefull when there are imbalanced classes. Since we have such a case (our Risk distribution is imbalanced) we select the f1-score as our metric.

### Cross Validation Analysis
We will select which classification methods to use by running a cross-validation on their base models. For this algorithm comparison we calculate the mean f1-scores and their standard deviation over a k-fold cross validation. Then, we use this score to select models for further analysis and optimization.

```python
# We add all models we learned about in the Machine Learning and Data Analytics class to a list
models = []
models.append(('LGR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('QDA', QuadraticDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('SVC', SVC(gamma='auto')))
```

```python
seed = 42
results = []
names = []
# select the f1-score as the metric
scoring = 'f1'

# perform a k-fold cross validation for every model
for name, model in models:
        cv = KFold(n_splits=10, shuffle=True, random_state=seed)
        cv_results = cross_val_score(model, x_train, y_train, cv=cv, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        # print results
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)

# calculate the mean f1-score
mean_score = np.mean([r.mean() for r in results])
print(f'Mean f1 score: {mean_score}')

# plot the results
fig = plt.figure(figsize=(11,6))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
ax.set_ylabel('F1 Score')
ax.boxplot(results)
ax.axhline(y=mean_score, color='r', linestyle='-.', linewidth=0.8, label='average score')
ax.set_xticklabels(names)
ax.legend()

plt.show()
```

It can be observed, that Logistic Regression, Linear Discriminant Analysis, Random Forest and Support Vector Classification perform best. Since Logistic Regression and Linear Discriminant Analysis often perform similarly, we will select Logistic Regression, Random Forest and Support Vector Classification for model optimization.

## 5. Model Optimization <a name='model_optimization'></a>
In this chapter we will tune our selected models by applying statistical analysis, subset selection and parameter tuning.

### Logistic Regression

#### Statistical Analysis
Firstly, we will use the statsmodel library to perform logistic regression and take a closer look at the resulting model

```python
# import the library
import statsmodels.api as sm
```

```python
# adding constant to predictors
x_train_logit = sm.add_constant(x_train)
x_test_logit = sm.add_constant(x_test)
```

```python
# creating a function for the logistic regression
def logistic_regression(x_train, x_test, y_train, y_test, summary=True):
    
    # creating and training the model
    logit = sm.Logit(y_train, x_train)
    results = logit.fit()
    
    if summary:
        print(results.summary())
    
    # predicting values and and plotting confusion matrix
    y_predicted = results.predict(x_test)
    y_predicted = y_predicted > 0.5
    cf_matrix = metrics.confusion_matrix(y_test, y_predicted)
    
    ax = sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')
    ax.set_title('Confusion Matrix\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    ax.xaxis.set_ticklabels(['Risk bad','Risk good'])
    ax.yaxis.set_ticklabels(['Risk bad','Risk good'])
    plt.show()

    # printing some metrics
    print(f'recall: {metrics.recall_score(y_test, y_predicted)}')
    print(f'precision: {metrics.precision_score(y_test, y_predicted)}')
    print(f'accuracy: {metrics.accuracy_score(y_test, y_predicted)}')
    print(f'error rate: {1-metrics.accuracy_score(y_test, y_predicted)}')
```

```python
# calling the regression function on a model with all predictors
logistic_regression(x_train_logit, x_test_logit, y_train, y_test)
```

The current model contains 19 predictors. We can analyse their statistical relevance by taking a look at their respective p-values. A typical threshold is p-value < 0.05.
The attributes with p-value < 0.05 are: <br>
- Age
- Duration
- Housing_own
- Saving accounts_rich
- Checking account_moderate 
<br><br>
Let us perform another regression using only these five predictors.

```python
# keeping only the constant and the relevant predictors
x_test_logit = x_test_logit[['const', 'Age', 'Duration', 'Housing_own', 'Saving accounts_rich', 'Checking account_moderate']]
x_train_logit = x_train_logit[['const', 'Age', 'Duration', 'Housing_own', 'Saving accounts_rich', 'Checking account_moderate']]
```

```python
# calling the regression function again
logistic_regression(x_train_logit, x_test_logit, y_train, y_test, summary=False)
```

Removing the irrevelant predictors has improved our accuracy by increasing the recall while the precision remains nearly the same.

#### Best Subset Selection
Now that only five predictors remain, let us choose perform a best subset selection of these predictors. Since we have already removed the majority of our predictors, we do not expect a large improvement of performance. On the other hand, it should have a short runtime with only five predictors and may give us interesting results.

```python
# Preparing the cross validation model
seed = 42
x_test_logit = x_test[['Age', 'Duration', 'Housing_own', 'Saving accounts_rich', 'Checking account_moderate']]
x_train_logit = x_train[['Age', 'Duration', 'Housing_own', 'Saving accounts_rich', 'Checking account_moderate']]

model = LogisticRegression()
cv = KFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
# We save the subsets and their respective cv-scores in a list
total_results = [(None, 0),(None, 0),(None, 0),(None, 0),(None, 0)]

p = x_train_logit.shape[1]
for set_size in range(1, p+1):
    # create all possible combinations of the five predictors
    comb = combinations(x_train_logit.columns, set_size)
    for feature_set in comb:
        # calculate cross validation f1-score for each feature set
        x_restricted = x_train_logit[list(feature_set)].values
        cv_results = cross_val_score(model, x_restricted, y_train, cv=cv, scoring='f1')
        score = cv_results.mean()

        n_features = len(list(feature_set))
        # if the highest score is reaches for that dimension of features, we save the score and subset
        if score > total_results[n_features - 1][1]:
            total_results[n_features - 1] = (feature_set, score)
```

```python
# We prepare the results for a visualization
subset_size = [len(result[0]) for result in total_results]
subset_scores = [result[1] for result in total_results]
best_score = max(subset_scores)
best_score_index = subset_scores.index(best_score)
best_features = list(total_results[best_score_index][0])

# plot the results and mark the best score
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(subset_size, subset_scores, 'k', marker='o')
ax.plot(best_score_index+1, best_score, 'o', markersize=30, markeredgewidth=2, markerfacecolor="None",
         markeredgecolor='green')

ax.set_xlabel('Number of Predictors')
ax.set_ylabel('F1 Score of prediction')
plt.show()
```

```python
print(f'In logistic regression the highest f1 score of {best_score} is reached with the subset: \n {best_features}')
```

Therefore we will change the subset for logistic regression

```python
x_test_logit = x_test[['Age', 'Duration', 'Housing_own', 'Checking account_moderate']]
x_train_logit = x_train[['Age', 'Duration', 'Housing_own', 'Checking account_moderate']]
```

#### Hyperparameter tuning
Hyperparameter tuning is not as influential in logistic regression as other methods, but we can still try to improve our solution a litte. Two hyperparameters that can be tuned are the penalty and C-value. The penalty-hyperparameter specifies the norm of the penalty term that is used to deal with irrelevant predictors. An L1-regularization is known as Lasso Regression while an L2-regularization is known as Ridge Regression. The C-value controls the regularization strength. If C is small, then there is a high regularization of parameters, which is meant to avoid overfitting. If C is high, then the model becomes more flexible, but can have higher variance.

```python
# create a grid of parameters to examine for optimal performance
param_grid = {"penalty": [None, 'l1', 'l2'],
              "C": [100, 10, 1.0, 0.1, 0.01],
              }
```

```python
# run a grid search using the cross validation as defined earlier
grid_search = GridSearchCV(model, param_grid=param_grid, cv=cv, scoring='f1', verbose=4)
grid_search.fit(x_train_logit, y_train)
```

```python
print(f'Best parameters: {grid_search.best_params_}')
```

Finally, let us fit our logistic regression model with the new hyperparameters:

```python
lr = LogisticRegression(penalty='l2', C=0.1)
lr.fit(x_train_logit, y_train)
lr_predictions = lr.predict(x_test_logit)
print(f'F1 Score after tuning: {metrics.f1_score(y_test, lr_predictions)}')
```

Now we have improved the model performance both with the subset selection and hyperparameter tuning.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18, 8), gridspec_kw={'width_ratios': [1.3, 1]})
f1_scores = {}

# iterate over every model
for key, model in models.items():
    if key == 'Tuned LGR':
        x_test_subset = x_test_logit
    else:
        x_test_subset = x_test
        
    # predict the values based on the selected subset
    y_pred = model.predict(x_test_subset)
    # calculate the f1 score
    f1_score = np.round(metrics.f1_score(y_test, y_pred), 3)
    f1_scores[key] = f1_score
    # calculate the false positive rate and true positive rate for the ROC curve
    y_pred_proba = model.predict_proba(x_test_subset)[:,1]
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
    auc = np.round(metrics.roc_auc_score(y_test, y_pred_proba), 3)
    
    # plot the ROC curves of the tuned models
    if key.startswith('Tuned'):
        ax1.plot(fpr,tpr,label=f'{key}: AUC = {auc}')


ax1.set_title('ROC Curve')
ax1.set_ylabel('True Positive Rate')
ax1.set_xlabel('False Positive Rate')
ax1.legend(loc=4)

# plot the f1 scores of all models
bar_chart = ax2.bar(list(f1_scores.keys()), list(f1_scores.values()), width=0.5, color=['blue', 'blue','orange','orange','green','green'])
ax2.bar_label(bar_chart, label_type='edge', color='black', size=15)
ax2.set_title('Final Model Performance')
ax2.set_ylabel('F1 Score')
ax2.set_xlabel('Model')
ax2.set_ylim([0.5, 1])

plt.show()
```
